# How to embed image inside Jupyter Notebook?

One of the main issues with Jupyter Notebook is embedding images inside the notebook. I found a way that works, and probably we can embed all the images in the same way if you think it is a good solution. The drawback is long documentation cells because of those images, and we have to encode each image before embedding in the notebook. 

Please let me know if this is a good solution. 

## Pros:

*  Jupyter Notebooks will become stand alone
*  The size of image is configurable using html attributes for image (e.g. width="50%" )
*  The captions could be added easily
*  Can be exported as stand alone HTML file
*  The image can be downloaded by double clicking on it

## Cons
* We Need to encode each image using an encoder such as [this one](https://www.base64-image.de/)
* When editing the cell, the image is shown as a long encoded document
* If any character in encoded image section alters, the image will not be shown anymore
* It will not rendered as an image on GitHub
